```python
def reference(x):
    '''
    Input: A string
    Output: The string with references removed
    
    Example:
    Input: 'the{7}'
    Output: 'the'
    '''
    
    # your code here
    x=' '.join(x)
    print(x)
    y=x.split('{')
    print(y)
    z=[text for text in y if '}' not in text]
    print(z)
    return z
```
